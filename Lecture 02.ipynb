# **Algorithm Complexities**

Lecture 2 focuses on understanding the time complexities of algorithms and ranking them according to their growth rates in the worst case.

## **What is Time Complexity**

`Time Complexity` of an algorithm measures the amount of time an algorithm takes to run as a function of the size of the input usually denoted as n
